```python
import pandas as pd
# Make your own simple table
my_data = {
    'Fruit': ['Apple', 'Banana', 'Orange'],
    'Price': [50, 20, 30],
    'Color': ['Red', 'Yellow', 'Orange']
}

my_df = pd.DataFrame(my_data)
print(my_df)
```

```python
fruits_data = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango', 'Grapes'],
    'Price': [200, 80, 150, 300, 250],
    'Color': ['Red', 'Yellow', 'Orange', 'Yellow', 'Purple'],
    'Taste': [8, 7, 9, 10, 8]
}

fruits_df = pd.DataFrame(fruits_data)
print("Fruits DataFrame:")
print(fruits_df)
```

```python
# TODO: Save to CSV
fruits_df.to_csv('fruits_data.csv', index=False)
print("Saved to 'fruits_data.csv'")

# TODO: Load from CSV
loaded_fruits = pd.read_csv('fruits_data.csv')
print("\nLoaded from CSV:")
print(loaded_fruits)
```
